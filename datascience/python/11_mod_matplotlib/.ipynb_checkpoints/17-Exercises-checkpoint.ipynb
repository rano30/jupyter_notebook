### Exercise 1. 

Read Total Profit for all months and show it using a line plot with Month Number as your baseline

Total profit data provided for each month. Generated line plot must include the following properties:

1. X label name = Month Number
2. Y label name = Total profit

````python
# Exercise 1
````

````python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
import pandas as pd

x = pd.read_csv('./data/company_sales_data.csv')
xx = np.linspace(1,12)


plt.xlabel('Months')
plt.ylabel('Profit')
plt.plot( x['month_number'], x['total_profit']);
````

### Exercise 3.

Read all product sales data and show it using a multiline plot

Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

````python
# facecream 	facewash 	toothpaste 	bathingsoap 	shampoo 	moisturizer 	
x = pd.read_csv('./data/company_sales_data.csv')
xx = np.linspace(1, 12)


plt.xlabel('Months')
plt.ylabel('total sold')
plt.plot( x['month_number'], x['facecream'], color = 'red', label='facecream')
plt.plot( x['month_number'], x['facewash'], color='blue', label='facewash')
plt.plot( x['month_number'], x['toothpaste'], color='green', label='toothpaste')
plt.plot( x['month_number'], x['bathingsoap'], color='yellow', label='bathingsoap')
plt.plot( x['month_number'], x['shampoo'], color='purple', label='shampoo')
plt.plot( x['month_number'], x['moisturizer'], color='black', label='moisturizer')
plt.legend();
````

### Exercise 4.

Read toothpaste sales data of each month and show it using a scatter plot

### Exercise 5.

Read face cream and face wash product sales data and show it using the bar chart

The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

````python
x = pd.read_csv('./data/company_sales_data.csv')
xx = np.linspace(1, 12)
months = np.arange(len(x.month_number))
width=.5


plt.xlabel('Months')
plt.ylabel('units sold per month')
wash = plt.bar(months - width/2, "facewash", width=.4, data=x, label='facewash', color='green')
cream = plt.bar(months + width/2, "facecream", width=.4, data=x, label='facecream', color='blue')
plt.legend((wash, cream), ['facewash', 'facecream']);
````
